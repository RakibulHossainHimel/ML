```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Start browser
driver = webdriver.Chrome()

# Go to the Medex ACI page
driver.get("https://medex.com.bd/companies/2/aci-limited/brands?page=1")

# Wait a little to make sure the page loads
time.sleep(2)


for j in range(2,5):
    for i in range(1,11):
        
        # Select the first medicine element (brand name)
        name_div = driver.find_element(By.XPATH, '/html/body/main/div/section/div/div[2]/div[{j}]/a[{i}]/div/div[1]'.format(j=j,i=i))
        #medicine_name_n_type = medicine_name_element.text
        # Get span (form)
        medicine_type = name_div.find_element(By.TAG_NAME, 'span').text

        # Get the full text, then remove the span text to isolate the brand name
        full_text = name_div.text
        medicine_name = full_text.replace(medicine_type, '').strip()

        medicine_weight_element = driver.find_element(By.XPATH, '/html/body/main/div/section/div/div[2]/div[{j}]/a[{i}]/div/div[2]'.format(j=j,i=i))
        medicine_weight = medicine_weight_element.text

        medicine_generic_name_element = driver.find_element(By.XPATH, '/html/body/main/div/section/div/div[2]/div[{j}]/a[{i}]/div/div[3]'.format(j=j,i=i))
        medicine_generic_name = medicine_generic_name_element.text

        medicine_price_element = driver.find_element(By.XPATH, '/html/body/main/div/section/div/div[2]/div[{j}]/a[{i}]/div/div[4]'.format(j=j,i=i))
        medicine_price = medicine_price_element.text

        print( medicine_name, medicine_type,  medicine_weight, medicine_generic_name, medicine_price)

# Close the browser
driver.quit()
```
